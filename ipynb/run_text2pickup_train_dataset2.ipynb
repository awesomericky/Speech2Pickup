## Load Preprocessed Data

```
tpd = Text2Pickup_dataset2()
metadata_path = './drive/MyDrive/Speech2Pickup/dataset2/text2pickup_npz_train'
image_path = './drive/MyDrive/Speech2Pickup/dataset2/image_file'
img_idx, sen_len, text_inputs, pos_outputs, total_images = tpd.load_metadata_and_image(metadata_path, image_path)
```

```
restore_flag = 0
restore_epoch = 0
# raise ValueError('Check restore_flag and restore_epoch again!! If confident # this code')
"""
1) restore_flag == 0
- Restore encoder model weights only
- restore_epoch == 0

2) restore_flag == 1
- Restore total model weights
- restore_epoch != 0
"""
if restore_flag == 0 and restore_epoch != 0:
    raise ValueError('Check restore_flag and restore_epoch')
elif restore_flag == 1 and restore_epoch == 0:
    raise ValueError('Check restore_flag and restore_epoch')

wandb_config = {'batch_size': 8,
                'max_epoch': 170,
                'num_hg_Depth': 4,
                'dim_hg_feat': 256,
                'learning_rate': 0.00001,
                'dropout_rate': 0.0,
                'dataset': 2}

if restore_flag == 0:
    id = wandb.util.generate_id()
    print('Wandb id: {}'.format(id))
    with open('/content/drive/MyDrive/Speech2Pickup/dataset2_model/Text2Pickup_model/wandbID.txt', 'w') as f:
        f.write(str(id))
elif restore_flag == 1:
    id = '1ol1z9dg'
else:
    raise ValueError('Unsupported restore_flag')

wandb_run = wandb.init(project='Speech2Pickup', name='Text2Pickup_dataset2', id=id, resume="allow", config=wandb_config)

tpd.train(img_resize = 256, # must be fixed
      heatmap_resize = 64, # must be fixed
      batch_size = 8,
      max_epoch = 170,
      save_stride = 1,
      num_hg_Depth = 4, # good be fixed
      dim_hg_feat = 256, # good be fixed
      dim_rnn_cell = 256, # good be fixed
      restore_flag = restore_flag,
      restore_path = '/content/drive/MyDrive/Speech2Pickup/dataset2_model/Text2Pickup_model/model.ckpt',
      restore_epoch = restore_epoch,
      total_images = total_images,
      total_heatmaps_path = '/content/drive/MyDrive/Speech2Pickup/dataset2/heatmap/train',
      heatmap_load_batch_size = 80,
      train_text_inputs = text_inputs,
      train_sen_len = sen_len,
      train_img_idx = img_idx,
      train_pos_outputs = pos_outputs,
      learning_rate = 0.00001, # good be fixed
      model_save_path = '/content/drive/MyDrive/Speech2Pickup/dataset2_model/Text2Pickup_model/model.ckpt')
```
